```python
# coverting duration into minute
def convert(duration):
  if len(duration.split()) == 2:
    hours=int(duration.split()[0][:-1])
    minutes=int(duration.split()[1][:-1])
    return hours*60+minutes
  else:
    return int(duration[:-1])*60
```

```python
df['Duration']=df['Duration'].apply(convert)
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
print(x_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
x_train
```

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
sns.distplot(y_test-y_pred)
plt.show()
```

```python
from sklearn.metrics import mean_absolute_percentage_error,mean_squared_error
print("MAPE is :",mean_absolute_percentage_error(y_test,y_pred))
```

```python
print("MSE is :",mean_squared_error(y_test,y_pred))
```

```python
from sklearn.metrics import r2_score
print("R2_score is",r2_score(y_test,y_pred))
```

```python
# Randomized search cv
from sklearn.model_selection import RandomizedSearchCV

# no.of tress in random forest
n_estimators=[int(x) for x in np.linspace(start=100,stop=1500,num=12)]
max_features=['auto','sqrt']  # no.of features to consider in every split
```
